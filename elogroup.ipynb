```
url = 'https://drive.google.com/u/0/uc?id=1lVcE4adsDiyRFm2MTOSsR1D9x-AxN5cD&export=download'
response = urllib.request.urlopen(url)
data = response.read()
buffer = io.BytesIO(data)
temp_file_path = '/tmp/temp.parquet'
with open(temp_file_path, 'wb') as temp_file:
    temp_file.write(data)
df_venda = spark.read.parquet("file://" + temp_file_path)
```

```
df_venda = df_venda.select('ID_loja', 'ID_produto', 'ID_cliente', 'ID_tempo', 'Qtde_vendida', 'Receita_venda')
```

```
# Aplicando as transformações para converter a string em decimal monetário
df_venda = df_venda.withColumn("Receita_venda_Tratado", regexp_replace(col("Receita_venda"), "R\\$", ""))
df_venda = df_venda.withColumn("Receita_venda_Tratado", regexp_replace(col("Receita_venda_Tratado"), "\\.", ""))
df_venda = df_venda.withColumn("Receita_venda_Tratado", regexp_replace(col("Receita_venda_Tratado"), ",", "."))
df_venda = df_venda.withColumn("Receita_venda_Tratado", col("Receita_venda_Tratado").cast("decimal(18,2)"))
```

```
df_venda.createOrReplaceTempView("Vendas")
```

```
%sql
SELECT * FROM Vendas
```

## 1 - Quantidade vendida no período entre 01/04/2020 a 01/04/2021;

```
%sql
SELECT Vendas.Qtde_vendida, 
       DATE_FORMAT(TO_DATE(Tempo.Data), 'dd/MM/yyyy') AS Data_Venda
FROM   Vendas
INNER  JOIN Tempo ON Tempo.Id_Tempo = Vendas.Id_Tempo
WHERE  Tempo.Data BETWEEN '2020-04-01' AND '2021-04-01'   
```

## 2 - Quantidade de clientes únicos do sexo feminino e estado civil divorciada

```
%sql
SELECT COUNT(DISTINCT CLI.Id_Cliente) Quantidade
FROM   Cliente CLI
WHERE  CLI.Sexo = 'Feminino' AND CLI.EstadoCivil = 'Divorciada' 
```

## 3 - Buscar os produtos com código 55,120,142 e que tiveram receita maior que R$120,00;

```
%sql
SELECT  P.COD_Produto
       ,P.NM_Produto
       ,V.Qtde_vendida
       ,V.Receita_venda
FROM   Vendas V
INNER  JOIN Produto P ON P.COD_Produto = V.ID_produto
WHERE  V.ID_produto IN (55, 120, 142)
       AND CAST(REPLACE(REPLACE(V.Receita_venda_Tratado, 'R$ ', ''), ',', '.') AS DECIMAL(18, 2)) > 120.00
```

## 4 - Verificar a receita da loja com o código 32 no período igual ou superior a 01/04/2021;

```
%sql
SELECT  L.COD_Loja
       ,L.NM_Loja
       ,V.Receita_venda
       ,DATE_FORMAT(TO_DATE(T.Data), 'dd/MM/yyyy') AS Data_Venda
FROM   Loja l 
INNER  JOIN Vendas V ON V.ID_Loja  = L.ID_Loja
INNER  JOIN Tempo  T ON T.Id_Tempo = V.Id_Tempo  
WHERE  l.COD_Loja = 32
       AND T.Data >= '2021-04-01'
```
